```python
home_df = df.groupby(['Season','HomeTeam'])['FTHG'].sum().reset_index()
```

```python
home_df.rename(columns={'HomeTeam' : 'Team' , 'FTHG' : 'TotalGoals'})
```

```python
away_df = df.groupby(['Season' , 'AwayTeam'])['FTAG'].sum().reset_index()
away_df.columns = ['Season' , 'Team' , 'TotalGoals']
```

```python
away_df
```

```python
season_df = pd.concat([home_df , away_df])
```

```python
season_df = season_df.groupby(['Season' , 'Team'])['TotalGoals'].sum().reset_index()
```

```python
season_df
```

```python
totalGoalsdf = season_df.groupby(['Team'])['TotalGoals'].sum().reset_index()
```

```python
totalGoalsdf.describe()
```

```python
totalGoalsdf = totalGoalsdf[totalGoalsdf['TotalGoals'] > 500]
```

```python
totalGoalsdf
```

```python
filtered_df = season_df[season_df['Team'].isin(totalGoalsdf['Team'])]
```

```python
filtered_df
```

```python
pivot_df = filtered_df.pivot(index = 'Season' , columns = 'Team' , values = 'TotalGoals').fillna(0)
```

```python
pl.figure(figsize = (20,20))
pivot_df.plot(kind = 'line' , marker = 'o')

pl.title('Goals scored per season')
pl.xlabel('Season')
pl.ylabel('Total Goals')
pl.legend(title = 'Team')
pl.grid(True)

pl.savefig('goalsPerSeason')
pl.show()
```

```python
teamApp = {}

for season in df['Season'].unique():
    
    homeApp = df.loc[df['Season'] == season , 'HomeTeam'].unique()
    awayApp = df.loc[df['Season'] == season , 'AwayTeam'].unique()
    
    teamsApp = set(homeApp).union(set(awayApp))
    
    teamApp[season] = list(teamsApp)
    
```

```python
print(teamApp)
```

```python
teamPresent = {}

for season , teams in teamApp.items():
    for team in teams:
        if team in teamPresent:
            teamPresent[team] += 1
        else:
            teamPresent[team] = 1
```

```python
print(teamPresent)
```

```python
teamAppearance = pd.DataFrame(teamPresent.items() , columns=['Team' , 'Appearances']).sort_values('Appearances' , ascending = False)
```

```python
teamAppearancedf = teamAppearance[teamAppearance['Team'].isin(totalGoalsdf['Team'])]
```

```python
teamAppearancedf
```

```python
new_df = totalGoalsdf.merge(teamAppearancedf , on = 'Team')
```

```python
new_df['AvgGoals'] = new_df['TotalGoals']/new_df['Appearances']
```
